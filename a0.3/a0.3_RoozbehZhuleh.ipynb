## 1. Log and Exp: Proving Inverse Relationship


```python
import numpy as np

# Show log(exp(x)) == x and exp(log(x)) == x for several values
x_vals = np.array([1, 2, 3, 4, 100, 2000, 2.5, np.e])
print("x\tlog(exp(x))\texp(log(x))")
for x in x_vals :
    print(f"{x:.2f}\t{np.log(np.exp(x)):.2f}\t\t{np.exp(np.log(x)):.2f}")

# Edge case: try negative and zero values (show error/NaN for log)
y_vals = np.array([0, -1, 5])
print("\nlog(exp(y)) for 0 and negative values:")
print(np.log(np.exp(y_vals)))
```

## 2. Plotting Sine Functions Variations



```python
import matplotlib.pyplot as plt

# x = np.arange(-2*np.pi, 2*np.pi,0.1)
x = np.linspace(-2*np.pi, 2*np.pi, 200)

# look at the description of each function and call the proper function from numpy
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.sin(2*x), label='sin(2x)')
plt.plot(x, np.sin(x/2), label='sin(x/2)')
plt.plot(x, np.sin(x)**2, label='sin¬≤(x)')
plt.legend()
plt.title("Variations of Sine Functions")
plt.show()
```

## 4. Features of the Sigmoid Function

```python

def sigmoid(x):
    return 1/(1+(np.e ** -x))

x = np.linspace(-10, 10, 200)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("œÉ(x)")
plt.grid(True)
plt.show()

print("Range:", (y.min(), y.max()), "| At x=0:", sigmoid(0))
```

### üëÄüîé What do you see?

#### **your answer:** a chart showing the Sigmoid function with a domain of {-10,10} and a range of {-1,1}.

## 5. Derivatives of Famous Functions

```python
x1 = np.linspace(-3, 3, 400)
plt.plot(x1, x1**2, label="x¬≤")
plt.plot(x1, np.gradient(x1**2, x1), '--', label="d/dx x¬≤")

x2 = np.linspace(0, 3, 400)
plt.plot(x2, np.exp(x2), label="exp(x)")
plt.plot(x2, np.gradient(np.exp(x2), x2), '--', label="d/dx exp(x)")

x3 = np.linspace(0.1, 3, 400)
plt.plot(x3, np.log(x3), label="log(x)")
plt.plot(x3, np.gradient(np.log(x3), x3), '--', label="d/dx log(x)")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

## 6. Gradient of Selected Functions

```python
# search and learn more about sumpy
import sympy as sp

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the function
f = x**2 * y + sp.sin(y)

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Function: f(x, y) =", f)
print("Partial derivative w.r.t x (‚àÇf/‚àÇx):", df_dx)
print("Partial derivative w.r.t y (‚àÇf/‚àÇy):", df_dy)
```
